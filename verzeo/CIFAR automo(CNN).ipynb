```python
import tensorflow as tf 
(x_train,y_train), (x_test,y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
x_train.shape  #50000 are no. of images in x_train, 32*32 is image size,  3(depth of an image)(depth is 3 for a coloured image)
```

```python
print(x_test.shape)
print(y_train.shape)
y_test.shape
```

```python
#labels 
classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
```

```python
classes
```

```python
#NORMALIZATION
x_train =x_train/255
x_test = x_test/255
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
i = random.randint(1,50000)
plt.imshow(x_train[i])
classes[y_train[i][0]]
```

```python
#ANN input, hidden, output
#input (32*32*3)3072 neurons
#hidden 2038 neurons
#output 10 neurons
from tensorflow import keras
from keras.layers import Dense
from keras.layers import Flatten
ann = keras.Sequential()
ann.add(Flatten(input_shape = (32,32,3)))
ann.add(Dense(2038, activation='relu'))
ann.add(Dense(10,activation='softmax'))
```

```python
ann.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history=ann.fit(x_train,y_train,epochs=10)
```

```python
from tensorflow import keras
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dropout
cnn = keras.Sequential()
```

```python
cnn.add(Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)))   #adding 32(3X3 filters)
cnn.add(MaxPooling2D((2,2)))  #2x2 window size

cnn.add(Conv2D(64,(3,3),activation='relu',input_shape=(32,32,3)))   #adding 64(3X3 filters)
cnn.add(MaxPooling2D((2,2)))  #2x2 window size

cnn.add(Conv2D(128,(3,3),activation='relu',input_shape=(32,32,3)))   #adding 128(3X3 filters)
cnn.add(MaxPooling2D((2,2)))  #2x2 window size

#ANN archi. from here

cnn.add(Flatten())  #96 input neurons
cnn.add(Dense(64, activation='relu'))
cnn.add(Dense(10, activation='softmax'))
```

```python
cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'])
history = cnn.fit(x_train, y_train, epochs=10)
```

```python
plt.plot(history.history['loss'])
```

```python
plt.plot(history.history['accuracy'])
```

```python
cnn.summary()
```

```python
y_pred=cnn.predict(x_test)
classes_x = np.argmax(y_pred,axis=1)
classes_x
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
