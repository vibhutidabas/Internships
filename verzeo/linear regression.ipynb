```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Salary_Data.csv')
df
```

```python
df.info()
```

```python
x=df.iloc[:,[0]].values
y=df.iloc[:,1].values
```

```python
#0. split the data into input and output
#1. call the algo
#2. fit the model
#3. predict new values by using the created model
#4. check accuracy
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
y
```

```python
model.predict([[9.2]])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,y_pred, c='r')
plt.show()
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error,r2_score
print(mean_squared_error(y,y_pred))
print(r2_score(y,y_pred)*100)
```
