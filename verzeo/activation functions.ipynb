```python
#relu-->if the no. is +ve the output will be no. itself otherwise 0

import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-3,3,33)

def relu(x):
  #return x*(x>0)
  return 1*(x>0)#derivative of relu
y=relu(x)
plt.plot(x,y)
plt.show()
```

```python
#sigmoid-->converts -infinity and info. to 1 and 0

x=np.linspace(-10,10,100)
s = 1/(1+np.exp(-x))
plt.plot(x,s,label='Sigmoid')
plt.legend()
plt.show()
```

```python
x=np.linspace(-10,10,100)
s=tf.nn.sigmoid(x)
plt.plot(x,s)
plt.show()
```

```python
#binary
#multiclass
#gradient descent

#binary cross entropy measure of the diff. between 2 probabli.
#(p*log(q))+((1-p)*log(1-q)) -BCE
import numpy as np
from math import log
p=[1,1,1,1,0,0,0,0]
q=[0.8,0.75,0.7,0.5,0.3,0.3,0.4,0.1]
def bce(p,q):
  return -( (p*log(q))+((1-p)*log(1-q))   )

result = []
for i in range(len(p)):
  ce=bce(p[i],q[i])
  result.append(ce)

mean_ce = np.mean(result)
print(round(mean_ce,3))
```
